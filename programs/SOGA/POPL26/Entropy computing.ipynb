```python
from scipy.stats import norm, truncnorm
import numpy as np 
from scipy.special import kn
```

```python
def entropy(x, dist):
    x = np.linspace(-10,10,500)
    dt = x[1] - x[0]
    H = 0
    for xi in x:
        if dist(xi) > 0:
            H += dt*dist(xi)*np.log(dist(xi))
    return -H
```

```python
def density_prod(x, sigma1=1, sigma2=1):
    return kn(0, abs(x)/(sigma1*sigma2))/(np.pi*sigma1*sigma2)

def gauss_prod(x, sigma1=1, sigma2=1):
    return norm.pdf(x, np.sqrt(sigma1**2*sigma2**2))

x = np.linspace(-5,5,500)

print('Entropy of product of Gaussian:', entropy(x, density_prod))
print('Entropy of Gaussian with same moments:', entropy(x, gauss_prod))
```

```python
def density_prod(x, sigma1=1, sigma2=1):
    return kn(0, abs(x)/(sigma1*sigma2))/(np.pi*sigma1*sigma2)

def mix_of_prod(x, p=0.25, sigma1=1, sigma2=1, sigma3=np.sqrt(2), sigma4=1):
    return p*density_prod(x, sigma1, sigma2) + (1-p)*density_prod(x, sigma3, sigma4)

def gauss_prod(x, sigma1, sigma2):
    return norm.pdf(x, np.sqrt(sigma1**2*sigma2**2))
    
def mix_of_gauss(x, p=0.25, sigma1=1, sigma2=1, sigma3=np.sqrt(2), sigma4=1):
    return p*gauss_prod(x, sigma1, sigma2) + (1-p)*gauss_prod(x, sigma3, sigma4)

def entropy(x, dist):
    x = np.linspace(-10,10,500)
    dt = x[1] - x[0]
    H = 0
    for xi in x:
        if dist(xi) > 0:
            H += dt*dist(xi)*np.log(dist(xi))
    return -H

x = np.linspace(-5,5,500)

print('Entropy mixture of product:', entropy(x, mix_of_prod))
print('Entropy mixture of gaussians:', entropy(x, mix_of_gauss))
```
